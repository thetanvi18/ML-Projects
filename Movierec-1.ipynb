```python
#Import necessary libraries
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv('/content/drive/My Drive/ML Projects Datasets/1. moviesdataset.csv')
```

```python
movies.head(5)
```

```python
movies.isnull().sum()
```

```python
movies.fillna('',inplace=True)
movies.isnull().sum()
```

```python
movies.columns
```

```python
movies1 = movies[['id','title','genre','overview']]
movies1.head()
```

```python
count_vec = CountVectorizer(max_features=10000, stop_words='english')
count_vec
```

```python
vec=count_vec.fit_transform(movies1['genre']).toarray()
vec
```

```python
sim=cosine_similarity(vec)
sim
```

```python
def recommend(movie_title):
  try:
    index = movies1[movies1['title']==movie_title].index[0]
  except IndexError:
    print(f"Movie '{movie_title}' not found")
    return

  distance = sorted(list(enumerate(sim[index])), reverse=True, key=lambda vector:vector[1])
  print("Recommended Movies:")
  for i in distance[0:10]:
    print(movies1.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
